```python

def my_linfit(x,y):
    sum_x = 0.0
    sum_y = 0.0
    sq_sum_x = 0.0
    sum_xy = 0.0
    for i in range(len(y)):
        sum_y = float(y[i]) + sum_y
        sum_x = float(x[i]) + sum_x
        sq_sum_x = float(x[i]*x[i]) + sq_sum_x
        sum_xy = float(x[i]*y[i]) + sum_xy
        
    b = ((sq_sum_x*sum_y)-(sum_xy*sum_x))/((len(y)*sq_sum_x)-(sum_x*sum_x))
    a = ((sum_xy)-(b*sum_x)) / sq_sum_x
    
    return a, b


def main():
    import matplotlib.pyplot as plt
    import numpy as np
    
    fig = plt.figure()
    plt.xlabel('x: number of steps')
    plt.ylabel('y: distance in mm')
    plt.title('Press mouse M times and then stop the GUI')
    plt.axis([0,100,0,100])
    plt.grid(True)
    import matplotlib
    matplotlib.use('TkAgg')
    
    global ix, iy
    global coords
    
    coords = plt.ginput(50, mouse_add= 1, mouse_pop =2, mouse_stop= 3)
    temp_d = {}
    x_coords = []
    y_coords = []
    
    temp_d = dict(coords)
    x_coords = list(temp_d)
    y_coords = list(temp_d.values())
    
    plt.show()
    x_t = np.zeros(len(coords))
    y_t = np.zeros(len(coords))
    for idx in range(len(coords)-1):
        x_t[idx] = x_coords[idx]
        y_t[idx] = y_coords[idx]
        plt.plot(x_t[idx],y_t[idx],"k+")
    a, b = my_linfit(x_t,y_t)
    xp = np.arange(0,100,1)
    plt.plot(xp,a*xp+b,"g-")
    plt.show()
    
main()
```
